#A POINT ROBOT


## Persamaan Dasar Gerak Translasi dan Rotasi
\begin{eqnarray}
s &=& s_0 + vt \\
\theta &=& \theta_0 + \omega t
\end{eqnarray}


## Gerak Translasi dan Rotasi 2D

pointrobotmotion.png

Vektor $\mathbf{v}$ adalah menotasikan kecepatan translasi sebuah object $P$. Vektor $\mathbf{v}$ terdiri dari $v_x = \dot{x} = \dfrac{\text{d}x}{\text{d}t}$ kecepatan translasi pada sumbu $x$ dan $v_y = \dot{y} = \dfrac{\text{d}y}{\text{d}t}$ kecepatan translasi pada sumbu $y$, yang dapat dihitung dengan

\begin{eqnarray}
v_x &= v \cos \theta \\
v_y &= v \sin \theta
\end{eqnarray}

Notasi $\boldsymbol{\xi}$ adalah sebuah vektor untuk merepresentasikan kecepatan translasi dan kecepatan rotasi atau angular $\omega$. $\boldsymbol{\xi}$ disebut juga spatial velocity dari sebuah objek. 

\begin{eqnarray}
\boldsymbol{\xi} &=& [\mathbf{v}, \dot{\theta}]^{\mathsf{T}} \\ 
&=& [v_x, v_y, \omega]^{\mathsf{T}}
\end{eqnarray}

## Visualisasi Gerak ROBOT dengan Kecepatan Konstan

Visualisasi gerak objek $P$ untuk $0<t<10s$ jika

1.   Kecepatan gerak translasi objek $P$, $v=0.6$ m/s
2.   Kecepatan gerak angular objek $\omega=0.1$ rad/s
3.   Pose awal objek $P$ terhadap frame reference adalah $\mathbf{p}^0 = [2,1]^{\mathsf{T}}$ dengan sudut $\theta_0 = 30^o$

dapat dilakukan denga program PYTHON berikut


```python
import numpy as np
import matplotlib.pyplot as plt

def rotation(theta):
  R = np.array([[np.cos(theta), -np.sin(theta)], [np.sin(theta), np.cos(theta)]])
  return R

def objectdrawing(o, d, R):
  i = np.array([1,0])
  ii = R @ i 
  oo = o + d
  plt.quiver(*oo, *ii, scale=1, color='b', scale_units='xy', angles = 'xy', width=0.005)
  plt.scatter(oo[0], oo[1], color='k')

def framedrawing(o, d, R):
  i = np.array([1,0])
  j = np.array([0,1])
  ii = R @ i 
  jj = R @ j 
  oo = o + d
  plt.quiver(*oo, *ii, scale=1, color='b', scale_units='xy', angles = 'xy', width=0.005)
  plt.quiver(*oo, *jj, scale=1, color='g', scale_units='xy', angles = 'xy', width=0.005)
  

#frame reference
R0 = rotation(0)
framedrawing(np.array([0,0]), np.array([0,0]), R0)

#pose awal
p = np.array([2.,1.])
theta = np.radians(30)
R = rotation(theta)
o = np.array([0,0])

#plot object pose awal
objectdrawing(o, p, R)

ts = 0.01
t = np.arange(0, 100, ts)
v = 0.6
omega = 0.1

px = [p[0]]
py = [p[1]]
theta_log= [theta]

for i in range(len(t)-1):
  theta = theta + omega * ts 
  p[0] = p[0] + v * np.cos(theta) * ts
  p[1] = p[1] + v * np.sin(theta) * ts
  px.append(p[0])
  py.append(p[1])
  theta_log.append(theta)


#plot object pose akhir
R = rotation(theta)
objectdrawing(o, p, R)

#plot trajectory
plt.plot(px, py)

plt.axis([ -15 ,6 , -1 ,15])
plt.grid(linestyle='-.', linewidth=0.5)
plt.xlabel("x(m)")
plt.ylabel("y(m)")
plt.show()

print("Posisi akhir = ", p)
print("Sudut orientasi akhir = ", np.degrees(theta))


#Plot trajectory perpindahan pose object
plt.figure(1)
plt.plot(t, px)
plt.xlabel("t(s)")
plt.ylabel("x(m)")
plt.grid(linestyle='-.', linewidth=0.5)

plt.figure(2)
plt.plot(t, py)
plt.grid(linestyle='-.', linewidth=0.5)
plt.xlabel("t(s)")
plt.ylabel("y(m)")

plt.figure(3)
plt.plot(t, np.degrees(theta_log)/360 * theta_log)
plt.grid(linestyle='-.', linewidth=0.5)
plt.xlabel("t(s)")
plt.ylabel(r"$\theta$" " (degrees)")
```

## An Exponential Decreases Error Controller

### Disain kontroller
Vektor $\mathbf{X} = [x,y,\theta]^{\mathsf{T}}$ merepresentasikan pose sebuah objek dalam ruang 2D. Jika $\mathbf{E}$ adalah sebuah vektor error antara pose robot saat $t$, $\mathbf{X}$ dan pose robot yang diinginkan $\mathbf{X}^*$ yaitu dituliskan

\begin{eqnarray}
\mathbf{E} = \mathbf{X} - \mathbf{X}^*
\end{eqnarray}

Turunan pertama, jika $\mathbf{X}^*$ adalah statis
\begin{eqnarray}
\dot{\mathbf{E}} = \dot{\mathbf{X}}
\end{eqnarray}

Fungsi exponential decreases
\begin{eqnarray}
\dot{\mathbf{E}} = -\lambda \mathbf{E}
\end{eqnarray}

Jadi didapatkan
\begin{eqnarray}
\dot{\mathbf{X}} &=& -\lambda \mathbf{E} \\ 
\boldsymbol{\xi} &=& -\lambda (\mathbf{X} - \mathbf{X}^*) \\ 
\boldsymbol{\xi} &=& \lambda (\mathbf{X}^* - \mathbf{X})
\end{eqnarray}

Elemen vektor $\boldsymbol{\xi}$

\begin{eqnarray}
\boldsymbol{\xi} &=& [\xi_0, \xi_1, \xi_2]^{\mathsf{T}} \\ 
&=& [\dot{x}, \dot{y}, \dot{\theta}]^{\mathsf{T}}
\end{eqnarray}

Menghitung pose robot 
\begin{eqnarray}
x(t) &=& x(t-1) + \Delta t \dot{x} \\ 
y(t) &=& y(t-1) + \Delta t \dot{y} \\
\theta(t) &=& \theta(t-1) + \Delta t \dot{\theta}
\end{eqnarray}

```python
import numpy as np
import matplotlib.pyplot as plt

def rotation(theta):
  R = np.array([[np.cos(theta), -np.sin(theta)], [np.sin(theta), np.cos(theta)]])
  return R

def objectdrawing(o, d, R):
  i = np.array([1,0])
  ii = R @ i 
  oo = o + d
  plt.quiver(*oo, *ii, scale=1, color='b', scale_units='xy', angles = 'xy', width=0.005)
  plt.scatter(oo[0], oo[1], color='k')

def framedrawing(o, d, R):
  i = np.array([1,0])
  j = np.array([0,1])
  ii = R @ i 
  jj = R @ j 
  oo = o + d
  plt.quiver(*oo, *ii, scale=1, color='b', scale_units='xy', angles = 'xy', width=0.005)
  plt.quiver(*oo, *jj, scale=1, color='g', scale_units='xy', angles = 'xy', width=0.005)
  

#frame reference
R0 = rotation(0)
framedrawing(np.array([0,0]), np.array([0,0]), R0)

#pose awal
p = np.array([2.,1.])
theta = np.radians(30)
X = np.array([p[0], p[1], theta])
R = rotation(theta)
o = np.array([0,0])

#pose target
#Xstar = np.array([4.99057855, 5.90857517,np.radians(87.23848373356597)])

Xstar = np.array([6, -2.5, np.radians(0)])

#Xstar = np.array([2,6,np.radians(180)])
#plot object pose awal
objectdrawing(o, p, R)

ts = 0.1
t = np.arange(0, 4, ts)

px = [p[0]]
py = [p[1]]
theta_log= [theta]
E = Xstar - X
E_log = [np.linalg.norm(E)]

gain = 6

for i in range(len(t)-1):
  Xi = gain * E
  X = X + Xi*ts
  E = Xstar - X
  E_log.append(np.linalg.norm(E))

  px.append(X[0])
  py.append(X[1])
  theta_log.append(X[2])


#plot object pose akhir
theta = X[2]
R = rotation(theta)
p[0] = X[0]
p[1] = X[1] 
objectdrawing(o, p, R)

#plot trajectory
plt.plot(px, py)

plt.axis([ -1 ,10 , -10 , 10])
plt.grid(linestyle='-.', linewidth=0.5)
plt.xlabel("x(m)")
plt.ylabel("y(m)")
plt.savefig("trajectory.pdf")
plt.show()

print("Posisi akhir = ", p)
print("Sudut orientasi akhir = ", np.degrees(theta))


#Plot trajectory perpindahan pose object
plt.figure(1)
plt.plot(t, px)
plt.xlabel("t(s)")
plt.ylabel("x(m)")
plt.grid(linestyle='-.', linewidth=0.5)
plt.savefig("xtrajectory.pdf")


plt.figure(2)
plt.plot(t, py)
plt.grid(linestyle='-.', linewidth=0.5)
plt.xlabel("t(s)")
plt.ylabel("y(m)")
plt.savefig("ytrajectory.pdf")


plt.figure(3)
plt.plot(t, np.degrees(theta_log))
plt.grid(linestyle='-.', linewidth=0.5)
plt.xlabel("t(s)")
plt.ylabel(r"$\theta$" " (degrees)")
plt.savefig("thetatrajectory.pdf")


print(np.shape(E_log))
plt.figure(4)
plt.plot(t, E_log)
plt.grid(linestyle='-.', linewidth=0.5)
plt.xlabel("t(s)")
plt.ylabel(r"$|\mathbf{E}|$")
plt.savefig("etrajectory.pdf")


```
