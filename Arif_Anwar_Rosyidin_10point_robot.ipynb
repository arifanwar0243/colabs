```python
import numpy as np
import matplotlib.pyplot as plt

def rotation(theta):
  R = np.array([[np.cos(theta), -np.sin(theta)], [np.sin(theta), np.cos(theta)]])
  return R

def framedrawing(o, d, R):
  i = np.array([1,0])
  j = np.array([0,1])
  ii = R @ i 
  jj = R @ j 
  oo = o + d
  plt.quiver(*oo, *ii, scale=1, color='b', scale_units='xy', angles = 'xy', width=0.005)
  plt.quiver(*oo, *jj, scale=1, color='g', scale_units='xy', angles = 'xy', width=0.005)

def vectordrawing(o, p):
  plt.quiver(*o, *p, scale=1, color='r', scale_units='xy', angles = 'xy', width=0.005)

#5 frames
# Diketahui
# p5
# d54, d43, d32, d21, d10
# theta54, theta 43, theta32, theta21, theta10
# Dicari p0

p10 = np.array([2, 5])
#translasi
d109 = np.array([3, 5])
d98 = np.array([2, 7])
d87 = np.array([4, 4])
d76 = np.array([-3, -2])
d65 = np.array([5, 6])
d54 = np.array([0, -1])
d43 = np.array([3, 9])
d32 = np.array([-5, -3])
d21 = np.array([-3, -7])
d10 = np.array([-2, -4])

#rotasi
theta109 = np.radians(12)
theta98 = np.radians(41)
theta87 = np.radians(65)
theta76 = np.radians(90)
theta65 = np.radians(42)
theta54 = np.radians(180)
theta43 = np.radians(80)
theta32 = np.radians(36)
theta21 = np.radians(31)
theta10 = np.radians(0)

R109 = rotation(theta109)
R98 = rotation(theta98)
R87 = rotation(theta87)
R76 = rotation(theta76)
R65 = rotation(theta65)
R54 = rotation(theta54)
R43 = rotation(theta43)
R32 = rotation(theta32)
R21 = rotation(theta21)
R10 = rotation(theta10)


p9 = R109@p10 + d109
p8 = R98@p9 + d98
p7 = R87@p8 + d87
p6 = R76@p7 + d76
p5 = R65@p6 + d65
p4 = R54@p5 + d54
p3 = R43@p4 + d43
p2 = R32@p3 + d32
p1 = R21@p2 + d21
p0 = R10@p1 + d10

#R10 = R10
#d10 = R10 + d10
R20 = R10@R21
d20 = R10@d21 + d10
R30 = R10@R21@R32
d30 = R10@R21@d32 + R10@d21 + d10
R40 = R10@R21@R32@R43
d40 = R10@R21@R32@d43 + R10@R21@d32 + R10@d21+ d10
R50 = R10@R21@R32@R43@R54
d50 = R10@R21@R32@R43@d54 + R10@R21@R32@d43 + R10@R21@d32 + R10@d21+ d10
R60 = R10@R21@R32@R43@R54@R65
d60 = R10@R21@R32@R43@R54@d65 + R10@R21@R32@R43@d54 + R10@R21@R32@d43 + R10@R21@d32 + R10@d21+ d10
R70 = R10@R21@R32@R43@R54@R65@R76
d70 = R10@R21@R32@R43@R54@R65@d76 + R10@R21@R32@R43@R54@d65 + R10@R21@R32@R43@d54 + R10@R21@R32@d43 + R10@R21@d32 + R10@d21+ d10
R80 = R10@R21@R32@R43@R54@R65@R76@R87
d80 = R10@R21@R32@R43@R54@R65@R76@d87 + R10@R21@R32@R43@R54@R65@d76 + R10@R21@R32@R43@R54@d65 + R10@R21@R32@R43@d54 + R10@R21@R32@d43 + R10@R21@d32 + R10@d21+ d10
R90 = R10@R21@R32@R43@R54@R65@R76@R87@R98
d90 = R10@R21@R32@R43@R54@R65@R76@R87@d98 + R10@R21@R32@R43@R54@R65@R76@d87 + R10@R21@R32@R43@R54@R65@d76 + R10@R21@R32@R43@R54@d65 + R10@R21@R32@R43@d54 + R10@R21@R32@d43 + R10@R21@d32 + R10@d21+ d10
R100 = R10@R21@R32@R43@R54@R65@R76@R87@R98@R109
d100 = R10@R21@R32@R43@R54@R65@R76@R87@R98@d109 + R10@R21@R32@R43@R54@R65@R76@R87@d98 + R10@R21@R32@R43@R54@R65@R76@d87 + R10@R21@R32@R43@R54@R65@d76 + R10@R21@R32@R43@R54@d65 + R10@R21@R32@R43@d54 + R10@R21@R32@d43 + R10@R21@d32 + R10@d21+ d10
origin = np.array([0,0])


#plt.figure(figsize=(10,10))
#frame 0
framedrawing(origin, origin, rotation(0))
#frame 1
framedrawing(origin, d10, R10)
#frame 2
framedrawing(origin, d20, R20)
#frame 3
framedrawing(origin, d30, R30)
#frame 4
framedrawing(origin, d40, R40)
#frame 5
framedrawing(origin, d50, R50)
#frame 6
framedrawing(origin, d60, R60)
#frame 7
framedrawing(origin, d70, R70)
#frame 8
framedrawing(origin, d80, R80)
#frame 9
framedrawing(origin, d90, R90)
#frame 10
framedrawing(origin, d100, R100)

#p0
vectordrawing(origin, p0)
#p1
vectordrawing(d10,  R10 @ p1)
#p2
vectordrawing(d20,  R10 @ R21 @ p2)
#p3
vectordrawing(d30,  R10 @ R21 @ R32 @ p3)
#p4
vectordrawing(d40,  R10 @ R21 @ R32 @ R43 @p4)
#p5
vectordrawing(d50,  R10 @ R21 @ R32 @ R43 @R54 @p5)
#p6
vectordrawing(d60,  R10 @ R21 @ R32 @ R43 @R54 @R65 @p6)
#p7
vectordrawing(d70,  R10 @ R21 @ R32 @ R43 @R54 @R65 @R76 @p7)
#p8
vectordrawing(d80,  R10 @ R21 @ R32 @ R43 @R54 @R65 @R76 @R87 @p8)
#p9
vectordrawing(d90,  R10 @ R21 @ R32 @ R43 @R54 @R65 @R76 @R87 @R98 @p9)
#p10
vectordrawing(d100,  R10 @ R21 @ R32 @ R43 @R54 @R65 @R76 @R87 @R98 @R109 @p10)



plt.text(0, 0, "0", fontsize = 8)
plt.text(*d10, '1', fontsize = 8)
plt.text(*d20, '2', fontsize = 8)
plt.text(*d30, '3', fontsize = 8)
plt.text(*d40, '4', fontsize = 8)
plt.text(*d50, '5', fontsize = 8)
plt.text(*d60, '6', fontsize = 8)
plt.text(*d70, '7', fontsize = 8)
plt.text(*d80, '8', fontsize = 8)
plt.text(*d90, '9', fontsize = 8)
plt.text(*d100, '10', fontsize = 8)


plt.axis([ -25 ,10 , -30 ,10])
plt.grid(linestyle='-.', linewidth=0.5)
plt.show()

print("p0 = ", p0)
print("p1 = ", p1)
print("p2 = ", p2)
print("p3 = ", p3)
print("p4 = ", p4)
print("p5 = ", p5)
print("p6 = ", p6)
print("p7 = ", p7)
print("p8 = ", p8)
print("p9 = ", p9)
print("p10 = ", p10)
print("d10 = ", d10)
print("d20 = ", d20)
print("d30 = ", d30)
print("d40 = ", d40)
print("d50 = ", d50)
print("d60 = ", d60)
print("d70 = ", d70)
print("d80 = ", d80)
print("d90 = ", d90)
print("d100 = ", d100)
#print("R10 @ R21 @ p2 = ", R10 @ R21 @ p2)
```
